```python
dfIdf = df.loc[df["libelle_regions"] == "Île-de-France"]
```

```python
dfIdfYearlyConsumption = dfIdf[["annee", "consototale"]].groupby(by="annee").sum().sort_values(by="annee", ascending=False)
dfIdfYearlyConsumption = dfIdfYearlyConsumption.reset_index()
```

```python
dfIdfYearlyConsumption
```

```python
import seaborn as sns
```

## Energy consumption of IDF through years

```python
sns.lineplot(data=dfIdfYearlyConsumption, x="annee", y="consototale")
```

```python
#df = df.groupby(by=["libelle_regions", "annee", "consototale"]).sum()
df["annee"].unique()
```

```python
# select only interested column
df2 = df[["annee", "libelle_regions", "consototale"]]
```

```python
dfConsYearNRegion = df2.groupby(by=["consototale"]).sum()
dfConsYearNRegion = dfConsYearNRegion.reset_index()
dfConsYearNRegion
```

## List of all regions and islands energy consumption from 2014 to 2022

```python
sns.lineplot(data=dfConsYearNRegion, x="annee", y="consototale", hue="libelle_regions")
```

```python
df["libelle_regions"].unique()
```

```python
# select only the regions in france
selected_regions = ["Auvergne-Rhône-Alpes", "Grand Est", "Occitanie", "Normandie", "Nouvelle-Aquitaine", "Centre-Val de Loire", "Corse", "Bourgogne-Franche-Comté", "Bretagne"]
filtered_dfConsYearNRegion = dfConsYearNRegion[dfConsYearNRegion['libelle_regions'].isin(selected_regions)]

sns.lineplot(data=filtered_dfConsYearNRegion, x="annee", y="consototale", hue="libelle_regions")
```
