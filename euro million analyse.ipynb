```python
import pandas
```

```python
# Calcul euro de fevrier 2020 à octobre 2024
# Dataset available here: https://www.fdj.fr/jeux-de-tirage/euromillions-my-million/historique
euromilliontable = pandas.read_csv("./Data/euromillions_202002.csv", delimiter=";")
```

```python
euro = euromilliontable[["boule_1", "boule_2", "boule_3", "boule_4", "boule_5"]]
```

```python
euro
```

```python
# concatenated rows into one

numberlist = pandas.concat([euro["boule_1"], euro["boule_2"], euro["boule_3"], euro["boule_4"], euro["boule_5"]])
numberlist.reset_index(drop=True)
```

```python
frequencies = dict(zip(list(range(1, 51)), [0]*50))
```

```python
def countOccurences(value):
    frequencies[value] = frequencies[value] + 1

numberlist.apply(countOccurences)
frequencies
```

```python
dfOccurences = pandas.DataFrame(frequencies.values(), columns=["value"], index=list(range(1, 51)))
```

```python
dfFrequencies = dfOccurences.sort_values(by="value", ascending=False) / 498
dfFrequencies.iloc[0:25]
```

```python
dfFrequencies.to_csv("./Data/euro_frequencies.csv")
```
